```python
import tika
from tika import parser
import re
class ResumeParsing:
    def parseAnyFile():
        tika.initVM()
        parsed = parser.from_file("resume.docx")
        print(parsed["content"])
        name=re.compile(r"([a-zA-Z]){3,}\s([a-zA-Z]){3,}")
        r2=re.search(name,parsed["content"])
        print(r2)
        regular_expression = re.compile(r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}", re.IGNORECASE)
        ph=re.compile(r"(\(\d{3}\) (\d{3}-\d{4}))|(\d{10})")
        r1=re.search(ph,parsed["content"])
        print(r1)
        result =re.search(regular_expression,parsed["content"])
        print(result)
        address =re.compile(r"Address:[0-9a-zA-Z\s]+[a-zA-Z]+,[a-zA-Z]+,[A-Z]*\s[0-9]*")
        r3=re.search(address,parsed["content"])
        print(r3)
    if __name__=="__main__":
        parseAnyFile()
```
